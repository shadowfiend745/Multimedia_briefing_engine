```python
from dotenv import load_dotenv
load_dotenv()
import anthropic
from pypdf import PdfReader
```

```python
filePath = "C:/Github/Multimedia_briefing_engine/Arknights_Endfield_Wikipedia.pdf"
reader = PdfReader(filePath)
rawText = " ".join("".join(page.extract_text() for page in reader.pages).split())
print(rawText[:500])  # Print the first 500 characters of the extracted text
```

```python
client = anthropic.Anthropic()

message = client.messages.create(
    model="claude-haiku-4-5-20251001",
    max_tokens=1024,
    system="""You are a content briefing assistant.
When given a piece of text, extract and structure the key ideas into:
- INTRO: One sentence explaining what this content is about
- KEY POINTS: 5 to 7 bullet points, each being one clear idea
- SUMMARY: Two sentences wrapping up the main takeaway

Reply in plain text using exactly those section headers.""",
    messages=[
        {
            "role": "user",
            "content": f"please extract the key ideas from the following text:\n\n{rawText}"
        }
    ]
)

briefing = message.content[0].text
print(briefing)
```

```python
with open("briefing.txt", "w") as f:    
    f.write(briefing)
```
