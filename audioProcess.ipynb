```python
from gtts import gTTS
from utilities import parser
import os
```

```python
intro, key_points, summary = parser("briefing.txt")

print("INTRO:")
print(intro)
print("\nKEY POINTS:")
for i, point in enumerate(key_points, start=1):
    print(f"{i}. {point}")
print("\nSUMMARY:")
print(summary)
```

```python
# gtts can only read string, so we need to join the key points into a single string
key_points_text = " ".join(key_points)
print("\nKEY POINTS TEXT:")
print(key_points_text)
```

```python
tts = gTTS(text=intro, lang='en')
tts.save("intro.mp3")
print("Intro audio saved as intro.mp3")

tts = gTTS(text=key_points_text, lang='en')
tts.save("key_points.mp3")
print("Key points audio saved as key_points.mp3")

tts = gTTS(text=summary, lang='en')
tts.save("summary.mp3")
print("Summary audio saved as summary.mp3")
```
