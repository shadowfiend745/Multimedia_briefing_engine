```python
from pypdf import PdfReader
from pptx import Presentation
from pptx.util import Inches, Pt
from pptx.dml.color import RGBColor
from utilities import add_textbox, parser
```

```python
intro = ''
key_point = ''
key_points = []
summary = ''
curr_section = ''
```

```python
intro, key_points, summary = parser("briefing.txt")

print("INTRO:")
print(intro)
print("\nKEY POINTS:")
for i, point in enumerate(key_points, 1):
    print(f"{i}. {point}")
print("\nSUMMARY:")
print(summary)
```

```python
prs = Presentation()
prs.slide_width = Inches(16)
prs.slide_height = Inches(9)
```

```python
intro_slide = prs.slides.add_slide(prs.slide_layouts[6])
add_textbox(intro_slide, 0.5, 0.5, 15, 1, "Arknights: Endfield", font_size=44, bold=True, italic=False, color="black")
add_textbox(intro_slide, 0.5, 2.0, 11, 2, intro,                 font_size=20, bold=False, italic=False, color="blue")
print("Intro slide created.")
```

```python
key_points_slide = prs.slides.add_slide(prs.slide_layouts[6])
add_textbox(key_points_slide, 0.5, 0.5, 15, 1, "KEY POINTS", font_size=44, bold=True, italic=False, color="black")

top = 1.5
for point in key_points:
    add_textbox(key_points_slide, 0.7, top, 11, 0.9, f"• {point}", font_size=20, bold=False, italic=False, color="blue")
    top += 1.0
print("Key points slide created.")
```

```python
summary_slide = prs.slides.add_slide(prs.slide_layouts[6])
add_textbox(summary_slide, 0.5, 0.5, 15, 1, "SUMMARY", font_size=44, bold=True, italic=False, color="black")
add_textbox(summary_slide, 0.5, 2.0, 11, 2, summary, font_size=20, bold=False, italic=False, color="blue")
print("Summary slide created.")
```

```python
prs.save("briefing_slides.pptx")
print("Presentation saved as briefing_slides.pptx")
```
